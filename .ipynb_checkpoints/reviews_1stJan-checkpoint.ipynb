```python
natural = spacy.load("en_core_web_sm")
eng_stop = stopwords.words('english')
def cleaning(reviews):
    list_clean=[]
    index = 0
    for review in tqdm(reviews):
        text = review.lower() # convert to lower case
        text = text.replace(" n't"," not") # n't changed to not
        text = text.replace("<br/>"," ") # break html tags removed
        text = text.replace("."," . ")
        text = text.replace(":)"," smile ")
        text = text.replace("_"," ")
                
        text = ' '.join(word.strip(string.punctuation) for word in text.split()) #strip punctuations
        
        words = text.split()
        words = [word for word in words if word.isalpha()] #remove numbers
        
        handled = [handle_appostrophe_dict2[word] if word in handle_appostrophe_dict2 else word for word in words] # handle_appostrophe_dict2 to handle apostrophe text
           
        handled = [word for word in handled if word not in eng_stop] # stop words removal
        processed = " ".join(handled)
        
        doc = natural(processed)
        processed = " ".join([token.lemma_ for token in doc]) #lemmatization
        list_clean.append(processed)
        index = index + 1
    df_cleaned = pd.DataFrame()
    df_cleaned['clean_reviews'] = list_clean
    return df_cleaned['clean_reviews']
X_cleaned = cleaning(X)
X_cleaned.head()

# References:
# - spaCy English language model: https://spacy.io/
# - English stop words: https://pythonspot.com/nltk-stop-words/
```
